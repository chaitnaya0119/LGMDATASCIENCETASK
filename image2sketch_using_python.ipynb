```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

**LOADING ORIGINAL IMAGE**

```python

image=cv2.imread('/content/dd1b999b911f86df02808fd75fe822510df8a961r1-1920-1080v2_uhq.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

**CONVERTING IMAGE TO GRAYSCALE**

```python
image_greyscale = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(image_greyscale,cmap="gray")
plt.axis('off')
plt.title('Gray scale image')
plt.show()
```

**Converting grey scale image to inverted image**

```python

image_invert = cv2.bitwise_not(image_greyscale)
plt.figure(figsize=(8,8))
plt.imshow(image_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

**Converting inverted image into smoothen image**

```python
img_smoothing = cv2.GaussianBlur(image_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

**Converting smoothen image into sketch**

```python
fin = cv2.divide(image_greyscale, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(fin,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
